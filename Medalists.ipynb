```python
import collections
```

```python
medal = collections.namedtuple('medal', ['year', 'athlete', 'team', 'event'])
medals = [medal(*line.strip().split('\t'))for line in open('goldmedals.txt', 'r')]
```

## Finds the five athletes who won the most golden medals

```python
top_athletes = collections.Counter(medal.athlete for medal in medals)
```
